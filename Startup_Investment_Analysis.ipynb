```python
"""
End-to-end Churn Prediction Pipeline (Fixed Version)
Includes:
- Synthetic dataset generation
- Feature engineering
- Train/test split
- RandomForestClassifier
- ELI5 permutation importance
- SHAP explainability
- Scoring & segmentation
- Export CSV

Required packages:
pip install pandas numpy scikit-learn eli5 shap matplotlib
"""

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from datetime import datetime, timedelta

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
from sklearn.inspection import permutation_importance

import eli5
from eli5.sklearn import PermutationImportance
import shap
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# -----------------------------
# 1) Synthetic data generation
# -----------------------------
def generate_synthetic_customers(n=5000):
    customer_id = np.arange(1, n+1)
    signup_dt = [datetime.now() - timedelta(days=int(x)) for x in np.random.exponential(400, n)]
    region = np.random.choice(['North','South','East','West'], size=n)
    plan = np.random.choice(['basic','plus','premium'], size=n)

    total_calls = np.random.poisson(40, n)
    total_call_minutes = np.maximum(0.5, np.random.normal(total_calls*2.5, 10))
    avg_call_min = np.divide(total_call_minutes, np.where(total_calls==0,1,total_calls))

    recharge_count = np.random.poisson(3, n)
    total_recharge_amt = recharge_count * np.random.choice([50,100,200,500], size=n)
    avg_recharge_amt = np.divide(total_recharge_amt, np.where(recharge_count==0,1,recharge_count))

    days_since_last_recharge = np.random.exponential(30, n).astype(int)
    days_since_last_call = np.random.exponential(25, n).astype(int)

    complaint_count = np.random.poisson(0.2, n)
    days_since_last_complaint = np.where(complaint_count>0, np.random.randint(1,90,size=n), np.nan)

    tenure_days = [(datetime.now()-d).days for d in signup_dt]

    df = pd.DataFrame({
        'customer_id': customer_id,
        'signup_dt': signup_dt,
        'region': region,
        'plan': plan,
        'total_calls': total_calls,
        'total_call_minutes': total_call_minutes,
        'avg_call_min': avg_call_min,
        'recharge_count': recharge_count,
        'total_recharge_amt': total_recharge_amt,
        'avg_recharge_amt': avg_recharge_amt,
        'days_since_last_recharge': days_since_last_recharge,
        'days_since_last_call': days_since_last_call,
        'complaint_count': complaint_count,
        'days_since_last_complaint': days_since_last_complaint,
        'tenure_days': tenure_days
    })

    # Churn label with simple rule + noise
    p = (
        0.003*df['days_since_last_recharge'] +
        0.002*df['days_since_last_call'] -
        0.4*(df['recharge_count']>0).astype(int) +
        0.6*df['complaint_count'] -
        0.0005*df['tenure_days']
    )
    churn_prob = 1/(1+np.exp(-p))
    churn_prob = 0.9*churn_prob + 0.1*np.random.rand(n)
    churn = (churn_prob > np.quantile(churn_prob, 0.78)).astype(int)

    df['churn'] = churn
    df['p_churn_true'] = churn_prob

    return df

df = generate_synthetic_customers(5000)

# -----------------------------
# 2) Modeling pipeline
# -----------------------------
Y = df['churn']
X = df.drop(columns=['customer_id','signup_dt','churn','p_churn_true'])

num_cols = X.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object','category']).columns.tolist()

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='constant', fill_value='MISSING')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preproc = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

pipe = Pipeline([
    ('preproc', preproc),
    ('clf', RandomForestClassifier(n_estimators=200, class_weight='balanced',
                                   random_state=RANDOM_STATE, n_jobs=-1))
])

X_train, X_test, y_train, y_test = train_test_split(X, Y, stratify=Y, test_size=0.2, random_state=RANDOM_STATE)

# Quick CV AUC
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring='roc_auc', n_jobs=-1)
print('Mean CV AUC: {:.4f}'.format(scores.mean()))

pipe.fit(X_train, y_train)

y_proba = pipe.predict_proba(X_test)[:,1]
y_pred = pipe.predict(X_test)
print('Test ROC AUC: {:.4f}'.format(roc_auc_score(y_test, y_proba)))

# -----------------------------
# 3) Explainability - Permutation Importance (ELI5)
# -----------------------------
# Get exact feature names
feature_names = pipe.named_steps['preproc'].get_feature_names_out()

# Run permutation importance on the full pipeline
perm = PermutationImportance(pipe, scoring='roc_auc', random_state=RANDOM_STATE, n_iter=10)
perm.fit(X_test, y_test)

# Show top features in notebook
eli5.show_weights(perm, feature_names=feature_names)

# Also build a DataFrame
imp_df = pd.DataFrame({
    'feature': feature_names,
    'importance': perm.feature_importances_
}).sort_values(by='importance', ascending=False)
print('\nTop 10 features by permutation importance:')
print(imp_df.head(10))

# -----------------------------
# 4) Explainability - SHAP
# -----------------------------
rf = pipe.named_steps['clf']
X_train_trans = pipe.named_steps['preproc'].transform(X_train)
X_test_trans = pipe.named_steps['preproc'].transform(X_test)

explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_train_trans)

# Global summary
shap.summary_plot(shap_values[1], X_train_trans, feature_names=feature_names, show=False)
plt.savefig('shap_summary.png')
plt.close()
print('Saved SHAP summary plot')

# Local waterfall plot for a sample customer
sample_idx = 3
sample_trans = pipe.named_steps['preproc'].transform(X_test.iloc[[sample_idx]])
sv = explainer.shap_values(sample_trans)
shap.plots._waterfall.waterfall_legacy(explainer.expected_value[1], sv[1][0], feature_names=feature_names, show=False)
plt.savefig('shap_waterfall_sample.png')
plt.close()
print('Saved SHAP waterfall for sample customer')

# -----------------------------
# 5) Scoring & segmentation
# -----------------------------
scored = df.copy()
scored['p_churn'] = pipe.predict_proba(X)[:,1]
scored['recharge_count_last_90'] = scored['recharge_count']

def assign_segment(row):
    if (row['p_churn'] >= 0.6) or (row['p_churn'] >= 0.4 and row['days_since_last_recharge'] > 30):
        return 'At Risk'
    if (row['days_since_last_recharge'] >= 90) and (row['recharge_count_last_90'] == 0):
        return 'Dormant'
    if (row['p_churn'] <= 0.2) and (row['recharge_count_last_90'] >= 2) and (row['tenure_days'] > 180):
        return 'Loyal'
    return 'Neutral'

scored['segment'] = scored.apply(assign_segment, axis=1)
scored[['customer_id','p_churn','segment','days_since_last_recharge','recharge_count_last_90','complaint_count']].to_csv('scored_customers.csv', index=False)
print('Saved scored_customers.csv')

print('\nSegment distribution:')
print(scored['segment'].value_counts())

top_at_risk = scored[scored['segment']=='At Risk'].sort_values('p_churn', ascending=False).head(10)
print('\nTop 10 At Risk customers:')
print(top_at_risk[['customer_id','p_churn','days_since_last_recharge','complaint_count']])
```

```python
# -----------------------------
# Complete Churn Pipeline with Explainability
# -----------------------------

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
from sklearn.inspection import permutation_importance

import eli5
from eli5.sklearn import PermutationImportance
import shap
import matplotlib.pyplot as plt

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# -----------------------------
# 1) Simulate SQL-like aggregated dataset
# -----------------------------
n = 1000
df = pd.DataFrame({
    'customer_id': np.arange(1, n+1),
    'signup_date': [datetime.now() - timedelta(days=int(x)) for x in np.random.exponential(400, n)],
    'region': np.random.choice(['North','South','East','West'], n),
    'plan': np.random.choice(['basic','plus','premium'], n),
    'total_calls': np.random.poisson(40, n),
    'total_call_minutes': np.random.normal(100, 30, n),
    'recharge_count': np.random.poisson(3, n),
    'total_recharge_amt': np.random.choice([50,100,200,500], n),
    'complaint_count': np.random.poisson(0.2, n),
    'days_since_last_call': np.random.randint(0, 100, n),
    'days_since_last_recharge': np.random.randint(0, 100, n),
    'days_since_last_complaint': np.random.randint(0, 100, n)
})

# Churn label: simple rule-based for demo
df['churn'] = ((df['days_since_last_recharge']>60) | (df['complaint_count']>1)).astype(int)

# -----------------------------
# 2) Features + target
# -----------------------------
X = df.drop(columns=['customer_id','signup_date','churn'])
y = df['churn']

num_cols = X.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object']).columns.tolist()

# -----------------------------
# 3) Preprocessing + pipeline
# -----------------------------
num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='constant', fill_value='MISSING')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

clf_pipeline = Pipeline([
    ('preproc', preprocessor),
    ('clf', RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=RANDOM_STATE))
])

# -----------------------------
# 4) Train/test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=RANDOM_STATE)

clf_pipeline.fit(X_train, y_train)

# Evaluate
y_proba = clf_pipeline.predict_proba(X_test)[:,1]
y_pred = clf_pipeline.predict(X_test)
print('ROC AUC:', roc_auc_score(y_test, y_proba))
print(classification_report(y_test, y_pred))
print('Confusion matrix:\n', confusion_matrix(y_test, y_pred))

# -----------------------------
# 5) Feature explainability - ELI5 Permutation Importance
# -----------------------------
# Get preprocessed feature names
feature_names = clf_pipeline.named_steps['preproc'].get_feature_names_out()

# Permutation importance on the classifier
perm_result = permutation_importance(
    clf_pipeline.named_steps['clf'], 
    clf_pipeline.named_steps['preproc'].transform(X_test),
    y_test,
    n_repeats=10,
    random_state=RANDOM_STATE,
    scoring='roc_auc'
)

# DataFrame for top features
imp_df = pd.DataFrame({
    'feature': feature_names,
    'importance': perm_result.importances_mean
}).sort_values(by='importance', ascending=False)

print('\nTop 10 features by permutation importance:')
print(imp_df.head(10))

# Optional: visualize using eli5
try:
    eli5.show_weights(PermutationImportance(clf_pipeline, random_state=RANDOM_STATE), feature_names=feature_names)
except:
    pass

# -----------------------------
# 6) SHAP explainability
# -----------------------------
rf_model = clf_pipeline.named_steps['clf']
X_train_trans = clf_pipeline.named_steps['preproc'].transform(X_train)
X_test_trans = clf_pipeline.named_steps['preproc'].transform(X_test)

explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_train_trans)

# Global summary plot
shap.summary_plot(shap_values[1], X_train_trans, feature_names=feature_names, show=False)
plt.title('SHAP Summary - Train Set')
plt.tight_layout()
plt.savefig('shap_summary.png')
plt.close()
print('\nSaved SHAP summary plot to shap_summary.png')

# Local explanation (sample customer)
sample_idx = 3
sample_row = X_test.iloc[[sample_idx]]
sample_trans = clf_pipeline.named_steps['preproc'].transform(sample_row)
sv = explainer.shap_values(sample_trans)

shap.plots._waterfall.waterfall_legacy(explainer.expected_value[1], sv[1][0], feature_names=feature_names, show=False)
plt.title('SHAP Waterfall - Sample Customer')
plt.tight_layout()
plt.savefig('shap_waterfall_sample.png')
plt.close()
print('Saved SHAP waterfall for sample customer to shap_waterfall_sample.png')

# -----------------------------
# 7) Add predicted churn + segmentation
# -----------------------------
df['p_churn'] = clf_pipeline.predict_proba(X)[:,1]

def assign_segment(row):
    if (row['p_churn'] >= 0.6) or (row['p_churn'] >= 0.4 and row['days_since_last_recharge'] > 30):
        return 'At Risk'
    if (row['days_since_last_recharge'] >= 90) and (row['recharge_count'] == 0):
        return 'Dormant'
    if (row['p_churn'] <= 0.2) and (row['recharge_count'] >= 2):
        return 'Loyal'
    return 'Neutral'

df['segment'] = df.apply(assign_segment, axis=1)
print('\nSegment distribution:')
print(df['segment'].value_counts())

# -----------------------------
# 8) Save scored dataset
# -----------------------------
df[['customer_id','p_churn','segment','days_since_last_recharge','recharge_count','complaint_count']].to_csv('scored_customers.csv', index=False)
print('\nSaved scored_customers.csv')
```
